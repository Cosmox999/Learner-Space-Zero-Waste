```python
def boundary_transversal_matrix(array):
  x = len(array)
  for i in range(x):
    print(arr[0][i])
  for i in range(1,x):
    print(arr[i][x-1])
  for i in range(1,x):
    print(arr[x-1][x-1-i])
  for i in range(1,x-1):
    print(arr[x-1-i][0])

print(arr)
boundary_transversal_matrix(arr)
```

# **Problem 2**

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
arr = np.random.randint(low=0, high=256, size=(20,20))
arr
```

```python
def max_min_median(array):

  for i in range(20):
    row_max = max(array[i])
    row_min = min(array[i])
    row_median = np.median(array[i])
    print(f"Row {i}: Max = {row_max}, Min = {row_min}, Median = {row_median}")

  for i in range(20):
    col_max = max(array[:][i])
    col_min = min(array[:][i])
    col_median = np.median(array[:][i])
    print(f"Col {i}: Max = {col_max}, Min = {col_min}, Median = {col_median}")

max_min_median(arr)
```

```python
img = cv2.resize(arr, (0,0), fx=10, fy=10, interpolation=cv2.INTER_NEAREST)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

```python
horizontal_gradient = np.tile(np.arange(256, dtype=np.uint8), (256, 1))
vertical_gradient = np.tile(np.arange(256, dtype=np.uint8).reshape(256, 1), (1, 256))
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Horizontal Gradient")
plt.imshow(horizontal_gradient, cmap='gray', vmin=0, vmax=255)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Vertical Gradient")
plt.imshow(vertical_gradient, cmap='gray', vmin=0, vmax=255)
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **Problem 3**

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Colab works/practice/dog image.jpeg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_gray, cmap='gray')
```

```python
def inversion(img):
  inv_img = np.zeros(shape=(img.shape[0], img.shape[1]), dtype=np.uint8)
  for i in range(img.shape[0]):
    for j in range(img.shape[1]):
      inv_img[i][j] = 255 - img[i][j]
  return inv_img

img_inv = inversion(img_gray)
plt.imshow(img_inv, cmap='gray')
```

```python

R, G, B = cv2.split(img_rgb)
R[:] = 0
modified_img = cv2.merge([R, G, B])
plt.imshow(modified_img)
```

```python
horizontal_flip_img = cv2.flip(img_rgb, 1)
vertical_flip_img = cv2.flip(img_rgb, 0)
mirror_img = horizontal_flip_img.copy()

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(horizontal_flip_img)
plt.title('Horizontally Flipped')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(vertical_flip_img)
plt.title('Vertically Flipped')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(mirror_img)
plt.title('Mirrored Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
def rgb2gray (img):
  R, G, B = cv2.split(img)
  gray_img = 0.299 * R + 0.587 * G + 0.114 * B
  return gray_img

img_gray = rgb2gray(img_rgb)
plt.imshow(img_gray, cmap='gray')
```
